```python
import pandas as pd
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
movies=pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits=credits.rename(columns={"movie_id":"id"})
```

```python
credits.head(1)
```

```python
movies.head(1)
```

```python
df=movies.merge(credits,on="id")
```

```python
df.head(1)
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(columns=["homepage","title_x","title_y","status","production_countries"],inplace=True)
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head(1)
```

```python
R=df["vote_average"]
v=df["vote_count"]
c=R.mean()
m=v.quantile(.70)
```

```python
df["weighted_avg"]=((R*v)+(c*m))/(v+m)
```

```python
df.head()
```

```python
df["weighted_avg"]=sorted(df["weighted_avg"],reverse=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
df[["normalised_weighted_avg","normalised_popularity"]]=ms.fit_transform(df[["weighted_avg","popularity"]])
```

```python
df.head()
```

```python
df["score"]=df["normalised_weighted_avg"]*.5 + df["normalised_popularity"]*.5
df["score"]=sorted(df["score"],reverse=True)
```
